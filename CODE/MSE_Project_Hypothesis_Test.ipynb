```
import pandas as pd
import numpy as np


numeric_columns = df.columns.difference(['State','STATE_NAME', 'Date'])


df[numeric_columns] = df[numeric_columns].replace({',': '', '\.': '', ' ': ''}, regex=True)
df[numeric_columns] = df[numeric_columns].replace('', np.nan).astype(float)

# Create the 'Unit Generated' column
df['Unit Generated'] = df['TOTAL (Sales)'] / df['TOTAL (Price)']

```

```
df['Unit_Generated']=df['Unit Generated'].round(2)
df.head(2)
```

```
df['Date'] = pd.to_datetime(df['Date'])
df['Day']=pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group by 'Month' and sum the 'Unit Generated' column for all states
df1 = df.groupby(['Year', 'Month','Day'])['Unit_Generated'].sum().reset_index()
df1['date']=pd.to_datetime(df['Date'])
```

Comparison of Two Groups: t-test

Null Hypothesis (H0): There is no significant difference in the mean 'Unit Generated' between two groups

Alternative Hypothesis (H1): There is a significant difference in the mean 'Unit Generated' between the groups.

```
df1.head(5)
```

```
from scipy.stats import ttest_ind

# Assume df1 is the DataFrame with 'Unit Generated' grouped by year and month
group1 = df1[(df1['Year'] >= 2009) & (df1['Year'] <= 2016)]['Unit_Generated']
group2 = df1[(df1['Year'] >= 2016) & (df1['Year'] <= 2023)]['Unit_Generated']

t_stat, p_value = ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, p-value: {p_value}")
```

t-statistic of approximately 5.09 and a very small p-value (4.696607939665177). The p-value is below conventional significance levels (e.g., 0.05), suggesting that you would reject the null hypothesis.

```
import pandas as pd
from scipy.stats import ttest_ind

# Get unique state names
states = df['STATE_NAME'].unique()

# Initialize lists to store the results
results = []

# Perform t-test for each pair of states
for i in range(len(states)):
    for j in range(i + 1, len(states)):
        state1 = states[i]
        state2 = states[j]

        # Select data for state1 and state2 from the respective time periods
        group_2009_to_2016_state1 = df[(df['STATE_NAME'] == state1) & (df['Year'] >= 2009) & (df['Year'] <= 2016)]['Unit_Generated']
        group_2016_to_2023_state1 = df[(df['STATE_NAME'] == state1) & (df['Year'] >= 2016) & (df['Year'] <= 2023)]['Unit_Generated']

        group_2009_to_2016_state2 = df[(df['STATE_NAME'] == state2) & (df['Year'] >= 2009) & (df['Year'] <= 2016)]['Unit_Generated']
        group_2016_to_2023_state2 = df[(df['STATE_NAME'] == state2) & (df['Year'] >= 2016) & (df['Year'] <= 2023)]['Unit_Generated']

        # Perform t-test for the two pairs of states
        t_statistic_state1, p_value_state1 = ttest_ind(group_2009_to_2016_state1, group_2016_to_2023_state1, equal_var=False)
        t_statistic_state2, p_value_state2 = ttest_ind(group_2009_to_2016_state2, group_2016_to_2023_state2, equal_var=False)

        # Store the results in a list
        results.append({
            'State 1': state1,
            'State 2': state2,
            'P-value': p_value_state1,
            'T-statistic': t_statistic_state1
        })
# Create a DataFrame from the results list
results_df = pd.DataFrame(results)
```
